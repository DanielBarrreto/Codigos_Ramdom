```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
def func(x):
  return x**2
def trans(p):
  return -1/4 * p**2
xx=sp.symbols("x")
def line(p):
  return trans(p)+p*xx
```

```python
x=np.linspace(-10,10,100)
x2=np.linspace(-20,20,100)
```

```python

for i in x2:
  pendiente=i
  ecu=line(pendiente)
  funcion=sp.lambdify(["x"],ecu)
  eje_y=funcion(x)
  plt.plot(x,eje_y)

proof=line(5)
plt.plot(x,func(x), color="k")
plt.ylim(-300,150)
```
